```python
from google.colab import files
import xlrd
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
uploaded = files.upload()
```

```python
data = pd.read_excel(uploaded['ConvertedTable.xlsx'])
data
```

```python
womenData = data[data['Пол'] == 1].copy()
menData = data[data['Пол'] == 0].copy()
#questionsData = data[data.columns[4:]].copy()
```

```python
#метод главных компонент (без учителя, показывается как метод кластеризации)

from sklearn.decomposition import PCA  # 1. Choose the model class
model = PCA(n_components=2)            # 2. Instantiate the model with hyperparameters
model.fit(womenData)                      # 3. Fit to data. Notice y is not specified!
X_2D = model.transform(womenData)         # 4. Transform the data to two dimensions
```

#GMM

```python
#смесь Гауссовых распределений (Gaussian mixture model, GMM)
from sklearn.mixture import GaussianMixture as GMM
modelGMM = GMM(n_components=3, covariance_type='full')
```

```python
#обучение только по женщинам
womenGMM = modelGMM.fit(womenData[womenData.columns[4:]].T)
womenY_GMM = womenGMM.predict(womenData[womenData.columns[4:]].T)
womenY_GMM
```

```python
#обучение только по мужчинам
menGMM = modelGMM.fit(menData[menData.columns[4:]].T)
menY_GMM = menGMM.predict(menData[menData.columns[4:]].T)
menY_GMM
```

#Kmeans


```python
#метод кластеризации к-средних
from sklearn.cluster import KMeans
modelKmeans = KMeans(n_clusters=3)
```

```python
#обучение только по женщинам
womenKmeans = modelKmeans.fit(womenData[womenData.columns[4:]].T)
womenY_Kmeans = womenKmeans.predict(womenData[womenData.columns[4:]].T)
womenY_Kmeans
```

```python
#обучение только по мужчинам
menKmeans = modelKmeans.fit(menData[menData.columns[4:]].T)
menY_Kmeans = menKmeans.predict(menData[menData.columns[4:]].T)
menY_Kmeans
```
